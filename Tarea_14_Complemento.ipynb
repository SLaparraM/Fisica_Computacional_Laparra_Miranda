# Ejercicio 4

Hacer lo mismo para el sistema: \begin{align} \dot x&= x+y-x (x^2+y^2)\ \dot y &= y-x-y(x^2+y^2) \end{align} Y comprobar numéricamente que hay un repulsor en $(0,0)$ y un ciclo límite (conjunto atractor) en $x^2+y^2=1$, para cotejar con lo que vimos en clase.

```julia
function RK4(f,list,x0)
     x = x0
     h = list[2]-list[1]
     listx=[]
     push!(listx,x)
     for i in 2:length(list)
        t=i*h
        k1=f(x,t);
        k2=f(x+(h/2)*k1,t+(h/2));
        k3=f(x+(h/2)*k2,t+(h/2));
        k4=f(x+h*k3,t+h);
        x = x + (h/6)*(k1+2*k2+2*k3+k4);
        push!(listx,x) 
    end
     return listx
end
```

```julia
f(x,t)=[x[1]+x[2]-x[1],x[2]-x[1]-x[2]]
```

```julia
list=0:0.01:100;
vec=RK4(f,list,[0,1]);
```

```julia
listx1=map(x->x[1],vec);
listx2=map(x->x[2],vec);
```

```julia
using PyPlot
```

```julia
xlabel("x")
ylabel("x'")
title("Espacio Fase (x(t),x'(t)) en x(0)=1")
grid("on")
plot(listx1,listx2) #Gràfica de la soluciòn aproximada con RK4 con una condiciòn inicial x(0)=1
```

```julia
list=linspace(0,10,1000)
for i in 1:100
    ci=4*rand(2,1)-2
    y=RK4(f,list,ci)
     xlabel("x")
    ylabel("x'")
title("Espacio Fase (x(t),x'(t))")
    grid("on")
    plot(map(x->x[1],y),map(x->x[2],y)) 
end
```
